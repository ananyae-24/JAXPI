```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x,y=np.array([.5,1,1.5]),np.array([.492,.98,1.488])
```

```python
plt.plot(x,y)
plt.xlabel("Actual Mean")
plt.ylabel("Predited Mean")
plt.title("Actual VS Predicted mean (at 0 Noise)")
```

```python
x,y=np.array([0,.5,1.5]),np.array([.581,.589,.621])
plt.plot(x,y)
plt.xlabel("Actual noise")
plt.ylabel("Predited Noise")
plt.title("Actual VS Predicted Noise (at .5 mean)")
```
